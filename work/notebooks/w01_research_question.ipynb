## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

I chose the CTR / Engagement Opportunity Scoring lane.

From the starter dataset, CTR changes a lot across ranking positions. Pages on page 1 have an average CTR of about 35.5%, while pages in deeper positions have an average CTR of only about 5.5%. This suggests that ranking position has a strong relationship with user engagement.

I want to build a system that helps identify pages with the best opportunity to improve CTR so that content teams can decide which pages to optimize first.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Decision

Which webpages should be prioritized for CTR optimization?

### Action

Content and SEO teams will review the selected pages and improve their titles, meta descriptions, content, and other on-page elements to increase click-through rate.

### Cost of a wrong call

If the wrong pages are chosen, the team may spend time optimizing pages that have little chance of improving CTR while overlooking pages with much greater potential. This can lead to missed opportunities for more organic traffic and inefficient use of resources.

### Why data or machine learning can help

The dataset contains several signals about page performance. Instead of relying on intuition, data can identify patterns that indicate which pages have the greatest opportunity for CTR improvement. A machine learning model can combine these signals to rank pages according to their optimization potential.

### Lane: CTR / Engagement Opportunity Scoring

Task Type: Scoring (Ranking)

The goal is to rank webpages by their optimization priority. Instead of making a yes/no prediction, the model assigns each webpage a priority score so editors know which pages should be reviewed first.

## 2. Target or proxy

The output is a priority score that ranks webpages by their likelihood of benefiting from content optimization.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

### Why ML Beats a Fixed Rule

Decision:
Which webpages should an editor optimize first?

Who acts?
A content editor or SEO specialist uses the ranked output to prioritize optimization work.

Cost of a wrong recommendation:
Editors may spend time improving low-impact pages while overlooking pages with greater potential to increase traffic.

Why ML?
A fixed rule based on only one feature, such as CTR or search volume, cannot capture the combined effect of multiple search and content signals. A scoring model can learn these patterns and produce a more useful ranking for decision-making.

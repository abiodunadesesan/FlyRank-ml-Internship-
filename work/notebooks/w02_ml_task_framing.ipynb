## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

### Lane: CTR / Engagement Opportunity Scoring

Task type: Scoring

I chose the CTR / Engagement Opportunity Scoring lane because the goal is to assign each webpage an opportunity score based on its likelihood of benefiting from optimization. Instead of making a simple yes-or-no prediction, the model scores and prioritizes webpages so that content teams know which pages to improve first. This supports better decision-making when there are many pages competing for limited time and resources.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

### Target or Proxy

The target is a CTR opportunity score for each webpage.

The starter dataset does not contain a direct "CTR opportunity" column, so a proxy target must be created from the available data. Pages with high impressions but relatively low CTR are likely to have the greatest opportunity for improvement.

The model will learn patterns from features such as impressions, clicks, CTR, ranking position, and other page characteristics to estimate which pages should be prioritized for optimization.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

### Success Metric

The success of the model will be measured by how well it identifies webpages with the highest opportunity for CTR improvement.

A useful metric is Precision@K, which measures how many of the top-ranked pages are genuinely good candidates for optimization.

The goal is to help content teams prioritize pages more effectively than using simple manual rules.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

### Unit of Analysis

The unit of analysis is one webpage.

Each row in the dataset represents a single webpage and contains information about its search performance, including impressions, clicks, CTR, ranking position, and other content-related features.

The model will assign a CTR opportunity score to each webpage so that pages with the greatest improvement potential can be prioritized.

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("/content/content_refresh_anonymized.csv")

# Show the unit of analysis (one row = one webpage)
unit_df = df[
    [
        "content_id",
        "client_id",
        "search_volume",
        "ctr",
        "avg_position",
        "engagement_rate",
        "word_count"
    ]
].head(10)

unit_df
```

```python
# Sketch a proxy target for CTR opportunity

df["ctr_opportunity"] = (
    (df["search_volume"] >= 100) &
    (df["ctr"] < df["ctr"].median())
)

df[
    [
        "content_id",
        "search_volume",
        "ctr",
        "ctr_opportunity"
    ]
].head(10)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

### Why ML Beats a Fixed Rule

A fixed rule, such as selecting every page with a CTR below 2%, ignores many other factors that influence optimization opportunities.

A machine learning approach can learn patterns from multiple features at the same time, including search volume, impressions, CTR, average position, engagement rate, and content characteristics. This allows the model to identify pages that have the highest potential for improvement instead of relying on one manually chosen threshold.

The output supports content teams by helping them prioritize which webpages to optimize first.
